# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6,0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

## pyplot



```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos", linestyle="--")
plt.xlabel("x")
plt.xlabel("y")
plt.title("sin & cos")
```

```python
plt.legend()
```

```python
plt.show()
```
